## 2. 测试基础连接

````python
import requests
import json
from datetime import datetime

def test_connection():
    """
    测试 AnythingLLM 连接
    """
    headers = {
        "Authorization": f"Bearer {ANYTHINGLLM_API_KEY}",
        "Content-Type": "application/json"
    }
    
    # 测试 1: 获取工作区列表
    try:
        url = f"{ANYTHINGLLM_BASE_URL}/api/v1/workspaces"
        resp = requests.get(url, headers=headers, timeout=10)
        print(f"✅ 连接成功！状态码: {resp.status_code}")
        
        if resp.status_code == 200:
            data = resp.json()
            print(f"工作区数量: {len(data.get('workspaces', []))}")
            for ws in data.get('workspaces', []):
                print(f"  - {ws.get('name', 'Unknown')} (ID: {ws.get('slug', 'N/A')})")
        return True
    except Exception as e:
        print(f"❌ 连接失败: {e}")
        return False

# 执行测试
test_connection()
````

## 3. 测试向量搜索 API

AnythingLLM 的搜索 API 格式：
- 端点：`/api/v1/workspace/{workspace_identifier}/vector-search` 或 `/search`
- 方法：POST
- ��数：`{"query": "搜索内容", "limit": 10}`

````python
def vector_search(query: str, limit: int = 10):
    """
    在 AnythingLLM 工作区中执行向量搜索
    
    Args:
        query: 搜索查询（如 "心理学书籍"）
        limit: 返回结果数量
    
    Returns:
        List[Dict]: 检索结果列表
    """
    headers = {
        "Authorization": f"Bearer {ANYTHINGLLM_API_KEY}",
        "Content-Type": "application/json"
    }
    
    # 尝试不同的端点格式
    endpoints = [
        f"{ANYTHINGLLM_BASE_URL}/api/v1/workspace/{WORKSPACE_IDENTIFIER}/vector-search",
        f"{ANYTHINGLLM_BASE_URL}/api/v1/workspace/{WORKSPACE_IDENTIFIER}/search",
        f"{ANYTHINGLLM_BASE_URL}/api/v1/workspace/{WORKSPACE_IDENTIFIER}/retrieve",
    ]
    
    payload = {"query": query, "limit": limit}
    
    for url in endpoints:
        try:
            print(f"尝试端点: {url}")
            resp = requests.post(url, headers=headers, json=payload, timeout=30)
            
            if resp.status_code == 200:
                data = resp.json()
                print(f"✅ 搜索成功！找到 {len(data.get('results', []))} 条结果")
                return data.get('results', [])
            else:
                print(f"  状态码: {resp.status_code}")
                print(f"  响应: {resp.text[:200]}")
        except requests.exceptions.RequestException as e:
            print(f"  请求失败: {e}")
            continue
    
    print("❌ 所有端点都失败，请检查 API 文档")
    return []

# 测试搜索
print("=" * 50)
print("测试向量搜索")
print("=" * 50)

results = vector_search("心理学书籍", limit=5)

print(f"\n检索到 {len(results)} 条结果")
for i, result in enumerate(results, 1):
    print(f"\n【结果 {i}】")
    print(f"相似度: {result.get('score', 'N/A')}")
    print(f"内容: {str(result.get('text', ''))[:200]}...")
    if 'metadata' in result:
        print(f"元数据: {json.dumps(result['metadata'], ensure_ascii=False)[:100]}...")
````

## 4. 解析检索结果

将 AnythingLLM 返回的文本片段解析为图书信息

````python
def parse_book_info(text: str) -> dict:
    """
    从文本片段中解析图书信息
    假设格式：
        书名：XXX
        作者：XXX
        ISBN：XXX
        ...
    """
    book = {
        'title': None,
        'author': '未知',
        'isbn': '未知',
        'category': '未分类',
        'description': ''
    }
    
    lines = text.split('\n')
    for line in lines:
        line = line.strip()
        if line.startswith('书名：') or line.startswith('书名:'):
            book['title'] = line.split('：', 1)[-1].strip()
        elif line.startswith('作者：') or line.startswith('作者:'):
            book['author'] = line.split('：', 1)[-1].strip()
        elif line.startswith('ISBN：') or line.startswith('ISBN:'):
            book['isbn'] = line.split('：', 1)[-1].strip()
        elif line.startswith('分类：') or line.startswith('分类:'):
            book['category'] = line.split('：', 1)[-1].strip()
        elif line.startswith('简介：') or line.startswith('简介:'):
            book['description'] = line.split('：', 1)[-1].strip()
    
    return book

# 测试解析函数
sample_text = """书名：心理学与生活
作者：理查德·格里格
ISBN：9787111574316
出版社：机械工业出版社
分类：心理学
简介：经典心理学教材"""

parsed = parse_book_info(sample_text)
print("解析结果:")
print(json.dumps(parsed, ensure_ascii=False, indent=2))
````

## 5. 完整测试：推荐和问答

模拟后端的推荐和问答逻辑

````python
def test_recommend(query: str, limit: int = 5):
    """
    测试推荐功能
    """
    print(f"\n{'='*50}")
    print(f"搜索推荐: {query}")
    print(f"{'='*50}")
    
    results = vector_search(query, limit=10)
    
    if not results:
        print("未找到相关图书")
        return []
    
    # 转换为推荐格式
    recommendations = []
    seen_titles = set()
    
    for result in results:
        text = result.get('text', '')
        book = parse_book_info(text)
        
        if book['title'] and book['title'] not in seen_titles:
            seen_titles.add(book['title'])
            recommendations.append({
                'book': book,
                'score': result.get('score', 0),
                'reason': '基于向量检索匹配'
            })
        
        if len(recommendations) >= limit:
            break
    
    print(f"\n推荐结果 ({len(recommendations)} 本):")
    for i, rec in enumerate(recommendations, 1):
        book = rec['book']
        print(f"\n{i}. 《{book['title']}》")
        print(f"   作者: {book['author']}")
        print(f"   分类: {book['category']}")
        print(f"   匹配度: {rec['score']:.4f}")
    
    return recommendations

# 测试推荐
test_recommend("我想读心理学相关的书")
````

````python
def test_qa(question: str):
    """
    测试问答功能
    """
    print(f"\n{'='*50}")
    print(f"提问: {question}")
    print(f"{'='*50}")
    
    results = vector_search(question, limit=10)
    
    if not results:
        print("未找到相关资料")
        return {
            'answer': '抱歉，未找到相关图书信息。',
            'recommendations': []
        }
    
    # 提取图书信息
    books_info = []
    context_parts = []
    
    for result in results[:5]:
        text = result.get('text', '')
        book = parse_book_info(text)
        
        if book['title'] and book not in books_info:
            books_info.append(book)
        
        context_parts.append(f"• {text[:200]}...")
    
    context = "\n".join(context_parts)
    
    # 生成回答
    answer = f"根据检索到的图书信息，关于「{question}」：\n\n"
    answer += f"找到了 {len(books_info)} 本相关图书：\n\n"
    
    for i, book in enumerate(books_info, 1):
        answer += f"{i}. 《{book['title']}》"
        if book['author'] != '未知':
            answer += f" - {book['author']}"
        answer += f" (分类: {book['category']})\n"
    
    answer += "\n您可以查看详情了解更多信息。"
    
    # 生成推荐列表
    recommendations = []
    for i, book in enumerate(books_info[:5]):
        recommendations.append({
            'book': book,
            'score': 1.0 - i * 0.1,
            'reason': '基于向量检索匹配'
        })
    
    print(f"\n回答: {answer}")
    print(f"\n推荐图书: {len(recommendations)} 本")
    
    return {
        'answer': answer,
        'recommendations': recommendations
    }

# 测试问答
result = test_qa("推荐一些心理学入门书籍")
````

## 6. 批量测试

测试多个查询，验证效果

````python
test_queries = [
    "编程入门书籍",
    "心理学经典著作",
    "历史小说推荐",
    "经济学入门"
]

for query in test_queries:
    test_recommend(query)
    print()
````
